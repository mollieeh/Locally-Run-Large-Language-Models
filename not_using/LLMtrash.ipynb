```python
import matplotlib.pyplot as plt

# Wrap the text before plotting (assuming format_text is defined)
def format_text(text, width=80):
    import textwrap
    return '\n'.join(textwrap.wrap(text, width))

# Group prompts and formatted outputs
prompts = [Q1_prompt, Q2_prompt, Q3_prompt, Q4_prompt, Q5_prompt]

outputs = [
    [Q1_output_tiny, Q1_output_deep, Q1_output_mist],
    [Q2_output_tiny, Q2_output_deep, Q2_output_mist],
    [format_text(Q3_output_tiny), format_text(Q3_output_deep), format_text(Q3_output_mist)],
    [format_text(Q4_output_tiny), format_text(Q4_output_deep), format_text(Q4_output_mist)],
    [format_text(Q5_output_tiny), format_text(Q5_output_deep), format_text(Q5_output_mist)],
]

titles = ["Tiny Llama", "DeepSeek", "Minstral"]
x_positions = [0.01, 0.40, 0.80]  # relative horizontal positions

# Plot each prompt and its 3-model output
for i in range(len(prompts)):
    # First, display the prompt
    fig, ax = plt.subplots(figsize=(15, 1.5))
    ax.text(0.01, 1.05, f'Prompt {i+1}: {prompts[i]}', fontsize=14, weight='bold')
    ax.axis('off')
    plt.show()

    # Then, show the outputs side by side
    fig, ax = plt.subplots(figsize=(15, 6))
    for x, title, text in zip(x_positions, titles, outputs[i]):
        ax.text(x, 1.05, title, fontsize=14, weight='bold')
        ax.text(x, 0.99, text, fontsize=12, va='top')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
```
